```python
import pandas as pd
from datetime import datetime
import os
import numpy as np
import pymongo
```

```python
appended_data = []
file_year = 2016
path = os.path.join("Original_Files","HRSAData","Annual_Data")
for WorkingFile in os.listdir(path):
        #identifies the list of files to be grouped together from designated folder
        print(WorkingFile)
        
        # Import the excel file by joining the path directory and file name together and call Ranked Data and Additonal Measure tabs
        file = pd.read_excel(os.path.join(path, WorkingFile),sheet_name=f'{file_year} Age and Race-Ethnicity',header=0)
        file2 = pd.read_excel(os.path.join(path, WorkingFile),sheet_name=f'{file_year} Patient Characteristics',header=0)
        file3 = pd.read_excel(os.path.join(path, WorkingFile),sheet_name=f'{file_year} Services',header=0)
        file4 = pd.read_excel(os.path.join(path, WorkingFile),sheet_name=f'{file_year} Clinical Data',header=0)
        file5 = pd.read_excel(os.path.join(path, WorkingFile),sheet_name=f'{file_year} Cost',header=0)
       
        # Load the excel files as two dataframes
        df = pd.DataFrame(file)
#         print(df.head(1))
        df2 = pd.DataFrame(file2)
#         print(df2.head(1))
        df3 = pd.DataFrame(file3)
#         print(df3.head(1))
        df4 = pd.DataFrame(file4)
#         print(df4.head(1))
        df5 = pd.DataFrame(file5)
#         print(df5.head(1))

        if file_year == 2018:
            df4 = df4.drop(['Depression Screening','Diabetes Control (Diabetic Patients with HbA1c > 9%)'],axis=1)
        else:
            df4= df4.drop(['Diabetes Control (Diabetic Patients with HbA1c <= 9%)'],axis=1)
        
        # Join two dataframes
        joined_df = pd.merge(df, df2, on='Health Center Name', how='outer')
        joined_df = joined_df.drop(['City_y','State_y'],axis=1)
        joined_df.rename(columns = {'State_x':'State','City_x':'City'},inplace = True)
        
        joined_df = pd.merge(joined_df, df3, on='Health Center Name', how='outer')
        joined_df = joined_df.drop(['City_y','State_y'],axis=1)
        joined_df.rename(columns = {'State_x':'State','City_x':'City'},inplace = True)
        
        joined_df = pd.merge(joined_df, df4, on='Health Center Name', how='outer')
        joined_df = joined_df.drop(['City_y','State_y'],axis=1)
        joined_df.rename(columns = {'State_x':'State','City_x':'City'},inplace = True)
        
        joined_df = pd.merge(joined_df, df5, on='Health Center Name', how='outer')
        joined_df = joined_df.drop(['City_y','State_y'],axis=1)
        joined_df.rename(columns = {'State_x':'State','City_x':'City'},inplace = True)
        
        joined_df['Year'] = file_year
        
        # append to list
        appended_data.append(joined_df)
        
        file_year += 1
        
#convert to dataframe
appended_data = pd.concat(appended_data)
```
