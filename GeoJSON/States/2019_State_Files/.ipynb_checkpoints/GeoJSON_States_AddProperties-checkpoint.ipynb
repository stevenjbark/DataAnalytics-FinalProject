```python
#Now every item in the list is a dictionary in the same order as the states_data dataframe from the working geojson file.
#Try replacing the "properties" dictionary with the dictionary from the test_list at the same index.
for i in range(len(health_list)):
    statesdata["features"][i]["properties"] = health_list[i]["properties"]
```

```python
#The dataframe statesdata gives a FeatureCollection in front of every line! This is NOT proper geojson format.
#The geojson format has all features ("features" column in dataframe) as a list of dictionaries. Try creating an empty
#list and then iterating through the dataframe and appending each line through .to_dict
statesdata_list = []

for row in statesdata.iterrows():
    statesdata_list.append(row[1][1])
```

```python
statesdata_list
```

```python
#Create final dictionary in the format of a geojson file with a FeatureCollection of features.
final_dict = {"type":"FeatureCollection", "features":statesdata_list}
```

```python
#Write final_dict to a geojson file
with open("2019_States_Health_Data.geojson","w") as file:
    json.dump(final_dict, file)
```
