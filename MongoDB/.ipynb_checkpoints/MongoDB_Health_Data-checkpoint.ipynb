# Uploading Data to Mongo Database

```python
#Open connection and create client through pymongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
#Create database (db) and collection for state data. Note, state and county data are appended to different
#collections in the same database.
db = client.ProjectHealth_DB
collection1 = db.StateData
collection2 = db.CountyData
```

```python
#Iterate through state_list and append each dictionary element into the StateData collection.
for data in state_list:
    collection1.insert_one(data)
```

```python
#Iterate through county_list and append each dictionary element into the CountyData collection.
for data in county_list:
    collection2.insert_one(data)
```
